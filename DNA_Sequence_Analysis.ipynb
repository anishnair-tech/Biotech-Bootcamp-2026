```python
# TRANSCRIPTION
# The starting DNA sequence
dna_sequence = "ATGCGACTATCGATCTAATCG"

# Transcribe DNA to RNA by replacing T with U
rna_sequence = dna_sequence.replace("T", "U")

print(f"Original DNA: {dna_sequence}")
print(f"Transcribed RNA: {rna_sequence}")
```

```python
# CODON DICTIONARY
# 64-codon universal genetic code table
codon_table = {
    'AUA':'I', 'AUC':'I', 'AUU':'I', 'AUG':'M', # Methionine / Start
    'GUA':'V', 'GUC':'V', 'GUG':'V', 'GUU':'V', # Valine
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCU':'A', # Alanine
    'GAA':'E', 'GAG':'E', 'GAC':'D', 'GAU':'D', # Glutamic Acid / Aspartic Acid
    'UUA':'L', 'UUG':'L', 'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L', # Leucine
    'UCA':'S', 'UCC':'S', 'UCG':'S', 'UCU':'S', # Serine
    'UAC':'Y', 'UAU':'Y', # Tyrosine
    'UAA':'_', 'UAG':'_', 'UGA':'_', # Stop Codons
    'UGC':'C', 'UGU':'C', 'UGG':'W', # Cysteine / Tryptophan
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGU':'R', # Arginine
    'AGU':'S', 'AGC':'S', 'AGA':'R', 'AGG':'R', # Serine / Arginine
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACU':'T', # Threonine
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGU':'G', # Glycine
    'CAA':'Q', 'CAG':'Q', 'CAC':'H', 'CAU':'H', # Glutamine / Histidine
    'AAU':'N', 'AAC':'N', 'AAA':'K', 'AAG':'K', # Asparagine / Lysine
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCU':'P', # Proline
    'UUU':'F', 'UUC':'F'                         # Phenylalanine
}

print(f"Success! Total codons in table: {len(codon_table)}")
```

```python
# TRANSLATION
#Split the full RNA sequence across lines for better readability without modifying the data
clean_rna = (
    "ACUGCUGACGAGAAGUCCGCCGUUACUGCCCUGUGGGGCAAGGUGAACGUGGAUGAAGUUGGUGGUGAGG"
    "CCCUGUGCAGGCUGCUGGUGGUCUACCCUUGGACCCAGAGGUUCUUUGAGUCCUUUGGGGAUCUGUCCAC"
    "UCCUGAUGCUGUUAUGGGCAACCCUAAGUGAAGGCUCAU"
)

# 2. Reset translation string
protein_sequence = ""

# 3. Loop through the RNA sequence in steps of 3
for i in range(0, len(clean_rna) - 2, 3):
    codon = clean_rna[i:i+3]  # Grab a 3-letter block

    # Look up the codon in the codon table
    amino_acid = codon_table.get(codon, '?')

    # Stop translating if a Stop Codon is hit
    if amino_acid == '_':
        print("Hit a Stop codon. Stopping translation.")
        break
        
    protein_sequence += amino_acid

print(f"Final Translated Protein: {protein_sequence}")
```

```python
# PROTEIN MOLECULAR WEIGHT
# Average molecular weight of each amino acid residue (in Daltons)
aa_weights = {
    'A': 71.08, 'R': 156.19, 'N': 114.10, 'D': 115.09, 'C': 103.14,
    'E': 129.12, 'Q': 128.13, 'G': 57.05, 'H': 137.14, 'I': 113.16,
    'L': 113.16, 'K': 128.17, 'M': 131.19, 'F': 147.18, 'P': 97.12,
    'S': 87.08, 'T': 101.11, 'W': 186.21, 'Y': 163.18, 'V': 99.13
}

total_weight = 0.0

# Loop through each amino acid in the generated protein
for aa in protein_sequence:
    if aa in aa_weights:
        total_weight += aa_weights[aa]

# Add 18.02 Da for the terminal water molecule (H2O) bound to a complete protein chain
if len(protein_sequence) > 0:
    total_weight += 18.02

print(f"Protein Sequence: {protein_sequence}")
# Formatting to 2 decimal places
print(f"Total Molecular Weight: {total_weight:.2f} Da") 
```

```python
# Simulating a FASTA file downloaded from the NCBI database
raw_fasta_data = """>NM_000518.5 Human Hemoglobin Subunit Beta (HBB)
ACTGCTGACGAGAAGTCCGCCGTTACTGCCCTGTGGGGCAAGGTGAACGTGGATGAAGTT
GGTGGTGAGGCCCTGGGCAGGCTGCTGGTGGTCTACCCTTGGACCCAGAGGTTCTTTGAG
TCCTTTGGGGATCTGTCCACTCCTGATGCTGTTATGGGCAACCCTAAGGTGAAGGCTCAT"""

# 1. Split the raw multi-line text into a list of individual lines
lines = raw_fasta_data.split("\n")

# 2. Extract the header descriptor (the very first item at index 0) to know what gene it is
header = lines[0]

# 3. Isolate all biological sequence rows from index 1 to the end 
sequence_lines = lines[1:]

# 4. Concatenate all sequence lines into a single, continuous string
clean_dna_sequence = "".join(sequence_lines)

# 5. Output the parsed results to verify success
print("--- HEADER FOUND ---")
print(header)
print("\n--- CLEANED DNA SEQUENCE ---")
print(clean_dna_sequence) 
```

```python
# 1. Transcribe the parsed DNA into RNA
rna_sequence = clean_dna_sequence.replace("T", "U")

# 2. Reset translation variables
protein_sequence = ""

# 3. Translate codons to find the final protein fragment
for i in range(0, len(rna_sequence) - 2, 3):
    codon = rna_sequence[i:i+3].upper()
    amino_acid = codon_table.get(codon, '?')
    
    if amino_acid == '_':
        print("Hit a Stop codon. Stopping translation.")
        break
    
    protein_sequence += amino_acid

# 4. Print out the results
print("--- TRANSLATION RESULTS ---")
print(f"Gene Source: {header}")
print(f"Resulting Protein: {protein_sequence}")
```

```python
# 1. Create an empty dictionary to store counts
aa_counts = {}

# 2. Loop through the variable: protein_sequence
for aa in protein_sequence:
    if aa in aa_counts:
        aa_counts[aa] += 1
    else:
        aa_counts[aa] = 1

print("--- AMINO ACID COMPOSITION ---")
for amino_acid in sorted(aa_counts.keys()):
    count = aa_counts[amino_acid]
    percentage = (count / len(protein_sequence)) * 100
    print(f"Amino Acid {amino_acid}: Count = {count}, Percentage = {percentage:.2f}%") 
```

```python
# 1. In the clean_dna_sequence, the GAG codon is at index 9 to 12
normal_codon = clean_dna_sequence[9:12] 

# 2. Change the middle 'A' to 'T' to create the Sickle Cell trait mutation (GTG)
# Take everything up to index 10, add "T", and join everything from index 11 onward
mutated_dna = clean_dna_sequence[:10] + "T" + clean_dna_sequence[11:]
mutated_codon = mutated_dna[9:12]

print("--- GENETIC MUTATION SIMULATION ---")
print(f"Original sequence segment: {clean_dna_sequence[:15]}")
print(f"Mutated sequence segment:  {mutated_dna[:15]}")
print(f"Normal Codon:  {normal_codon}")
print(f"Mutated Codon: {mutated_codon}")

# 3. Transcribe the mutated strand
mutated_rna = mutated_dna.replace("T", "U")

# 4. Translate the mutated RNA strand using the exact loop logic
mutated_protein = ""
for i in range(0, len(mutated_rna) - 2, 3):
    codon = mutated_rna[i:i+3].upper()
    amino_acid = codon_table.get(codon, '?')
    
    if amino_acid == '_':
        break
    mutated_protein += amino_acid

print(f"\nNormal Protein:  {protein_sequence}")
print(f"Mutated Protein: {mutated_protein}") 
```
