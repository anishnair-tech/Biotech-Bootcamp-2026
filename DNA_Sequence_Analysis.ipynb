```python
# 1. Define a target DNA sequence
seq = "ATGCGATCGATCGATC"

# 2. Count individual bases using Python's built-in .count() method
g_count = seq.count("G")
c_count = seq.count("C")

# 3. Calculate total sequence length
total_len = len(seq)

# 4. Math: Total GC percentage formula
gc_percentage = ((g_count + c_count) / total_len) * 100

# 5. Display the findings
print("G count:", g_count)
print("C count:", c_count)
print("GC Content Percentage:", gc_percentage, "%")
```

```python
# 1. Take the original DNA sequence and replace every 'T' with a 'U'
rna_seq = seq.replace("T", "U")

# 2. Display the original vs. the new copy
print("Original DNA Sequence:", seq)
print("Transcribed RNA Sequence:", rna_seq)
```

```python
# TRANSCRIPTION
# The starting DNA sequence
dna_sequence = "ATGCGACTATCGATCGATCG"

# Transcribe DNA to RNA by replacing T with U
rna_sequence = dna_sequence.replace("T", "U")

print(f"Original DNA: {dna_sequence}")
print(f"Transcribed RNA: {rna_sequence}")
```

```python
# CODON DICTIONARY
# A standard genetic code dictionary mapping RNA codons to Amino Acids
codon_table = {
    'AUA':'I', 'AUC':'I', 'AUU':'I', 'AUG':'M', # M is Start Codon
    'GUA':'V', 'GUC':'V', 'GUG':'V', 'GUU':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCU':'A',
    'GAA':'E', 'GAG':'E', 'GAC':'D', 'GAU':'D',
    'UUA':'L', 'UUG':'L', 'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L',
    'UCA':'S', 'UCC':'S', 'UCG':'S', 'UCU':'S', 'UAC':'Y', 'UAU':'Y',
    'UAA':'_', 'UAG':'_', 'UGA':'_', # '_' represents Stop Codons
    'UGC':'C', 'UGU':'C', 'UGG':'W',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGU':'R',
    'AGU':'S', 'AGC':'S', 'AGA':'R', 'AGG':'R',
    'CUA':'L', 'CUC':'L', 'CUG':'L', 'CUU':'L',
    'CAA':'Q', 'CAG':'Q', 'CAC':'H', 'CAU':'H',
    'AAU':'N', 'AAC':'N', 'AAA':'K', 'AAG':'K',
} 
# Check if the dictionary works by looking up the Start Codon
print("Dictionary loaded successfully! AUG maps to:", codon_table['AUG']) 
```

```python
# TRANSLATION
protein_sequence = ""

# Loop through the RNA sequence in steps of 3
for i in range(0, len(rna_sequence) - 2, 3):
    codon = rna_sequence[i:i+3] # Grab a 3-letter block
    
    # Look up the codon in the table, default to '?' if not found
    amino_acid = codon_table.get(codon, '?')
    
    # Stop translating if a Stop Codon is hit
    if amino_acid == '_':
        print("Hit a Stop codon. Stopping translation.")
        break
        
    protein_sequence += amino_acid

print(f"Final Translated Protein: {protein_sequence}") 
```

```python
# PROTEIN MOLECULAR WEIGHT
# Average molecular weight of each amino acid residue (in Daltons)
aa_weights = {
    'A': 71.08, 'R': 156.19, 'N': 114.10, 'D': 115.09, 'C': 103.14,
    'E': 129.12, 'Q': 128.13, 'G': 57.05, 'H': 137.14, 'I': 113.16,
    'L': 113.16, 'K': 128.17, 'M': 131.19, 'F': 147.18, 'P': 97.12,
    'S': 87.08, 'T': 101.11, 'W': 186.21, 'Y': 163.18, 'V': 99.13
}

total_weight = 0.0

# Loop through each amino acid in the generated protein
for aa in protein_sequence:
    if aa in aa_weights:
        total_weight += aa_weights[aa]

# Add 18.02 Da for the terminal water molecule (H2O) bound to a complete protein chain
if len(protein_sequence) > 0:
    total_weight += 18.02

print(f"Protein Sequence: {protein_sequence}")
# Formatting to 2 decimal places
print(f"Total Molecular Weight: {total_weight:.2f} Da") 
```
